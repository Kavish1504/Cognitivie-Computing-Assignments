**g. Add a new column that categorizes employees based on their performance rating:**

**i. Excellent for ratings >= 4.5**

**ii. Good for ratings >= 4.0 but < 4.5**

**iii. Average for ratings < 4.0**

```python
import pandas as pd
df=pd.read_csv('employee_data.csv')

def define_rating(rating): #function to define performance rating corresponding to particular rating
    if(rating>=4.5):
        return "Excellent"
    elif(rating>=4.0 and rating<4.5):
        return "Good"
    else:
        return "Average"
df['Performance Rating']=df['Rating'].apply(define_rating) #adding new column to the dataset
print(df)
```
