Q1 Ask the user to enter a value (e.g., M) for a mathematical function.

1 Generate x values from -10 to 10 using np.linspace().

Compute y values for:

1. y= M⋅x

2. y=M⋅sin(x)

2 Plot both functions in a single figure:

o Use different colors and line styles.

o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as p
M=int(input("Enter the value for M:"))
x=np.linspace(-10,10,50)
y1=M*x**2
y2=M*np.sin(x)
p.plot(x,y1,color='r',linestyle='dashed')
p.plot(x,y2,color='g',linestyle='solid')
p.legend()
p.grid(True)
p.title(f'Plots of y={M}x^2 and y={M}sin(x)')
p.show()
```

Q2. Create a dataset of five subjects and scores.

1 Convert it into a Pandas DataFrame.

2 Plot the scores using a Seaborn bar plot with:

o Different colors for each bar.

o Annotations on top of each bar.

o Title, axis labels, and grid

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as p
import seaborn as sns
subject=["AI","Cognitive","DBMS","Probability","DAA"]
scores=[90,87,88,97,85]
df=pd.DataFrame({"Subject":subject,"Scores":scores})
graph=sns.barplot(x='Subject',y='Scores',data=df,palette='viridis')
for bar in graph.patches:
  height=bar.get_height()
  x_coord=bar.get_x()+bar.get_width()/2
  graph.text(x_coord,height+2,f'{int(height)}')
p.title('Scores in Subjects')
p.legend()
p.grid(alpha=0.4)
p.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().

1 Create a 2x2 subplot layout displaying:

o Line plot showing cumulative sum.

o Scatter plot with random noise.

2 Customize with titles, axis labels, and grids.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as p
import seaborn as sns
np.random.seed(102317012)
data=np.random.randn(50)
fig,ax=p.subplots(2,2)
ax[0,0].plot(np.cumsum(data),color='b',marker='o')
ax[0,0].set_title('Line Plot Showing Cumulative Sum')
ax[0,0].grid(alpha=0.4)

ax[0,1].scatter(range(50),data+np.random.randn(50)*0.2,color='r')
ax[0,1].set_title('Scatter Plot')
ax[0,1].grid(alpha=0.4)
p.show()
```

Q4.  Download Data-set from the below link

https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv

Apply ‘seaborn’ library to do the following.

1. Read Total profit of all months and show it using a line plot.

2. Read all product sales data and show it using a multiline plot.

3. Plot bar chart for all the features/attributes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as p
import seaborn as sns
df=pd.read_csv("company_sales_data.csv")
sns.lineplot(data=df,x='month_number',y='total_profit',color='b',marker='o')
p.legend()
p.grid(alpha=0.4)
p.title('Line Plot')
p.show()

for column in df[1:-1]:
  sns.lineplot(data=df,x='month_number',y=column,color='b',marker='o')
p.legend()
p.grid(alpha=0.4)
p.title('Multiline Plot')
p.show()

p.bar(df['month_number'], df['total_profit'], label='Total Profit')
p.xlabel("Month")
p.ylabel("Values")
p.title("Bar Chart of Total Profit")
p.legend()
p.grid(alpha=0.4)
p.show()
```
